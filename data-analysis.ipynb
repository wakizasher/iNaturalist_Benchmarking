# Seed

## Seed 42

```python
df_blur42 = pd.read_csv('/kaggle/input/vlm-results/blurry_taxa_results_seed_42.csv')
df_human42 = pd.read_csv('/kaggle/input/vlm-results/human_taxa_results_seed_42.csv')
df_taxa42 = pd.read_csv('/kaggle/input/vlm-results/other_taxa_results_seed_42.csv')

print("blur dataframe shape:", df_blur42.shape)
print("blur columns:", df_blur42.columns.tolist())
print("\nHuman DataFrame shape:", df_human42.shape)
print("Human columns:", df_human42.columns.tolist())
print("\nTaxa DataFrame shape:", df_taxa42.shape)
print("Taxa columns:", df_taxa42.columns.tolist())
```

```python
df_blur42 = df_blur42.rename(columns={'Image_Name': 'image_name','Other_Taxa_Present': 'blurry_image'})
df_human42 = df_human42.rename(columns={'Image_Name': 'image_name','Humans_Present': 'humans_present'})
df_taxa42 = df_taxa42.rename(columns={'Image_Name': 'image_name','Other_Taxa_Present': 'other_taxa'})
```

```python
df_seed42 = pd.merge(df_blur42, df_human42, on='image_name', how='inner')
df_seed42 = pd.merge(df_seed42, df_taxa42, on='image_name', how='inner')
```

```python
df_seed42.head()
```

```python
seed_number = 42
def extract_flower_directory_name(image_name): 
    if image_name.startswith("Bellis_perennis"): 
        return "Bellis_perennis"
    elif image_name.startswith("Leucanthemum_vulgare"):
        return "Leucanthemum_vulgare"
    else: 
        return "Matricaria_chamomilla"
```

```python
df_seed42['flower_name_for_path'] = df_seed42['image_name'].apply(extract_flower_directory_name)
```

```python
df_seed42['file_path'] = df_seed42.apply(lambda row: f"E:\iNaturalist\test_images_200\seed_{seed_number}\{row['flower_name_for_path']}\{row['image_name']}", axis=1)
```

```python
df_seed42.head()
```
